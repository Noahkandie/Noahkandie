```python
# Data cleaning
# Validity
# Create a column for date from the given seperate  measures of time and date
df['x'] =df['year'].astype(str) + df['month'].astype(str).str.zfill(2) + df['day'].astype(str).str.zfill(2)
df['date']=pd.to_datetime(df['x'])
df.drop(columns=['x'],inplace=True)
df
```

```python
# Drop irrelevant columns
columns_drop=['Cars','Scheduled at','Displayed comment','Charging Status','Geo point','Status','Address', 'Charge Slots','year','minute','Subscription status']
df.drop(columns=columns_drop,axis=0,inplace=True)
df.info()
```

```python
# Selecting the data required
# city of paris , month of april
df_paris=df[(df['City']=='Paris')&(df['month']==4)&(df['Rental status']=='operational')]
df_paris[:4]
```

```python
df_paris.info()
```

```python
df_paris.head(3)
```

```python
#df_paris['Bluecar difference']=df_paris['['Bluecar counter'].diff()
#df_paris.head()
df_paris=df_paris.sort_values(by=['ID','date','hour'])
df_paris['diff_bluecar'] = df_paris.groupby(['ID'])['Bluecar counter'].diff()
df_paris[['ID','Bluecar counter','date','diff_bluecar']][:5]
```

```python
# fill missing value with 0
df_paris['diff_bluecar'].fillna(0,inplace=True)
df_paris.head()
```

```python
#confirming changes
df_paris.isnull().value_counts()
```

```python

# Analysis
# 1. Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018

df_paris[df_paris['diff_bluecar'] < 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```

```python


# 2. Identify the most popular hour of the day for picking a shared electric car (Bluecar) in the city of Paris over the month of April 2018

df_paris[df_paris['diff_bluecar'] > 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```
